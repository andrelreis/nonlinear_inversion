# An optimization using Levenberg-Marquardt method

## Importing libraries

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from visual import histeq
```

## Defining functions

### Rosenbrock function

```python
def rosenbrock_function(x,y):
    '''
    Calculate the rosenbrock map.   
    
    '''
    p1 = x
    p2 = y
    z = 100.*(p2-p1**2)**2 + (1.- p1)**2
    return z
```

```python
def rosenbrock(p):
    '''
    Calculate the rosenbrock function.   
    
    '''
    p1 = p[0]
    p2 = p[1]
    z = 100.*(p2-p1**2)**2 + (1.- p1)**2
    return z
```

```python
def rosenbrock_grad(p):
    '''
    Calculate the gradient of rosenbrock function.
    '''
    p1 = p[0]
    p2 = p[1]
    J1 = -400.*p1*(p2-p1**2) - 2.*(1.- p1)
    J2 = 200.*(p2-p1**2)
    J = np.array([J1,J2])
    return J
```

```python
def residual_ros(p):
    '''
    Calculate the residual of rosenbrock function.
    '''
    p1 = p[0]
    p2 = p[1]
    J1 = 10.*(p2-p1**2)
    J2 = (1.-p1)
    r = np.array([J1,J2])
    return r
```

```python
def sensitivity_ros(p):
    '''
    Calculate the sensitivity matrix of rosenbrock function.
    '''
    p1 = p[0]
    p2 = p[1]
    
    A11 = -20.*p1
    A12 = 10.
    A21 = -1.
    A22 = 0.
    A = np.array([A11,A12,A21,A22]).reshape(2,2)
    return A
```

```python
def hessian(y,s,H0):
    '''
    Calculate the hessian
    '''
    y = y.reshape(2,1)
    s = s.reshape(2,1)
    yTs = np.dot(y.T, s)
    if abs(yTs) <= 1e-10:
        H = H0
    else:
        rho = 1/yTs
        syT = np.dot(s, y.T)
        ysT = np.dot(y, s.T)
        I = np.identity(2)
        IrhosyT = I - rho*syT
        IrhoysT = I - rho*ysT
        IrhosyTH0 = np.dot(IrhosyT, H0)
        H = np.dot(IrhosyTH0, IrhoysT) + rho*np.dot(s, s.T)
    
    return H
```

## Parameters for calculating the rosenbrock function 

```python
Nx = 20
Ny = 20
```

```python
x = np.linspace(-5.,5.,num=Nx)
y = np.linspace(-10.,10.,num=Ny)
X,Y = np.meshgrid(x,y)
```

```python
Z = rosenbrock_function(X,Y) 
```

```python
Z_ros = histeq(Z).reshape(Nx,Ny)
```

## Visuatization of the Rosenbrock function 


```python
title_font = 22
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(10,8))
plt.contourf(X, Y, Z_ros, 50)
plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
plt.xlabel('x', fontsize = title_font)
plt.ylabel('y', fontsize = title_font)
plt.annotate('Rosenbrock Function', xy = (0.05, 0.93), xycoords = 'axes fraction', fontsize=20)

plt.show()
```

## Calculating the Quasi-Newton method for Rosenbrock

```python
itmax = 100
eps = 1e-10

beta = 0.4 
c1 = 1e-4
c2 = .9
alpha = 1.
m = 0.


p0 = np.array([2.,2.])

px = [p0[0]]
py = [p0[1]]

phi_it = []
it  = []

H0 = np.identity(2)
for i in range(itmax):
    ros0 = rosenbrock(p0)
    r0 = residual_ros(p0)
    phi0 = np.linalg.norm(r0)
    J0 = rosenbrock_grad(p0)
    
    dp0 = - np.dot(H0,J0)  
    p = p0 + alpha*dp0
        
    ros = rosenbrock(p)
    r = residual_ros(p)
    phi = np.linalg.norm(r)
    J = rosenbrock_grad(p)
    
    y_qn = J - J0
    s = p - p0
    
    while (phi - phi0) > (alpha*c1*np.sum(J0*dp0)) and abs(np.sum(J*dp0)) > abs(c2*np.sum(J0*dp0)):
        m += 1.
        alpha = beta**m
        p = p0 + alpha*dp0
        
        ros = rosenbrock(p)
        r = residual_ros(p)
        phi = np.linalg.norm(r)
        J = rosenbrock_grad(p)
        
        y_qn = J - J0
        s_qn = p - p0
    
    H = hessian(y_qn,s_qn,H0)
    
    H0 = H
    J0 = J
    r0 = r
    phi0 = phi
    ros0=ros
    p0 = p
    
    px.append(p0[0])
    py.append(p0[1])
    phi_it.append(phi)
    it.append(i)  
```

```python

title_font = 22
bottom_font = 16
saturation_factor = 1.
plt.close('all')
plt.figure(figsize=(10,8))
plt.contourf(X, Y, Z_ros, 50)
plt.plot(px,py,'w.-')
plt.colorbar(pad=0.01, aspect=40, shrink=1.0)
plt.xlabel('x', fontsize = title_font)
plt.ylabel('y', fontsize = title_font)
plt.annotate('Rosenbrock Function', xy = (0.05, 0.93), xycoords = 'axes fraction', fontsize=20)

plt.show()
```
